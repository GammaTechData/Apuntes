```python
import numpy as np
```

```python
titanic_df.Age = titanic_df.Age.fillna((titanic_df.Age.mean()))
```

```python
q1=np.percentile(titanic_df.Age, 25)

q3=np.percentile(titanic_df.Age, 75)
```

```python
iqr=q3-q1

# 3IQR

upper = q3 + 1.5*iqr

lower = q1 - 1.5*iqr
```

```python
titanic_df=titanic_df[titanic_df.Age<upper] 
titanic_df
```

```python
q1=np.percentile(titanic_df.Fare, 25)

q3=np.percentile(titanic_df.Fare, 75)
```

```python
iqr=q3-q1

# 3IQR

upper = q3 + 1.5*iqr

lower = q1 - 1.5*iqr
```

```python
titanic_df=titanic_df[titanic_df.Fare<upper] 
titanic_df
```

## 21. Estudia la colinealidad

```python
corr = titanic_df.corr(numeric_only=True)
mascara = np.triu(titanic_df,1)
mascara = np.triu(np.ones_like(corr, dtype=bool)) # generamos la máscara para la triangular superior
color_map = sns.diverging_palette(220, 20, as_cmap=True) # Paleta de colores
sns.heatmap(corr,  
            mask = mascara,
            cmap=color_map,
            square=True, #que los datos se vean como cuadrados
            linewidth=0.5, #ancho de línea
            vmax=1,
            vmin=0,
            cbar_kws={"shrink": .5
                     },# barra lateral
            annot=True #Anotacion, que te ponga el valor de la corr() dentro de la casilla
)
```
