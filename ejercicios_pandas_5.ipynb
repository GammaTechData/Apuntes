## 1. Leer el archivo CSV: wine_quality.csv

```python
wine_df = pd.read_csv('datasets/wine_quality.csv')
wine_df[wine_df.duplicated(keep=False)].sort_values('pH')
```

## 3. Calcular la media de la columna "alcohol"

```python
wine_df.alcohol.mean()
```

## 4. Contar la cantidad de vinos que tienen un pH superior a 3.5.

```python
lst=[]

for i in wine_df.pH:
    if i > 3.5:
        lst.append(i)
        
len(lst)
```

## 5. Filtrar los vinos con un pH superior a 3.5 y una calidad mayor o igual a 7

```python
dafr = wine_df[(wine_df.pH>3.5) & (wine_df.quality >= 7)]

dafr
```

## 6. Ordenar el DataFrame por la columna "alcohol" de forma descendente.

```python
wine_df.sort_values(by="alcohol",ascending=False)
```

## 7. Calcular la calidad promedio para cada tipo de vino ("red" y "white") (pista: group_by)

```python
import numpy as np
wine_df["color"] = wine_df.apply(lambda row: np.random.choice(["red", "white"]), axis=1)

calidad = wine_df["quality"].groupby(wine_df["color"]).mean()
calidad
```

## 8. Calcular el valor mximo de la columna "chlorides" para vinos rojos con una calidad de al menos 7.

```python
wine_df[(wine_df.quality <= 7) & (wine_df.color == 'red')].sort_values(by="chlorides",ascending=False).head(1)
```

## 9. Calcular el valor mnimo, mximo y promedio de la columna "fixed acidity" para vinos blancos.

```python
wine_df[(wine_df.color == 'white')]['fixed acidity'].describe().T
```
